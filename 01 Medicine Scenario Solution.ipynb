### Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

### 1. In the Drive folder you have a dataset 'Diabetes_data.csv', read the dataset into Pandas.

```python
df = pd.read_csv('Diabetes_data.csv')
df
```

---

### 2. The data is slopily gathered and you need to rename, correct, and handle missing values. Check to see if any of the columns have null/missing values and determine the best strategy for cleaning up the data (including not doing anything).

```python
missing_values = df.isnull().any(axis=1).sum()
print(missing_values) # There are 97 rows with missing values, in the following interpretation I am gonna ignore the empty values, 
# significantly small, if I delete those 97 rows, I will delete other data that would affect my results.
new_columns = ['Пациент_ИБ','Број на месеци', 'Хемоглобин 3 месечен', 'Инсулин_1/Таблети_0', 'Глукоза во крв', 'Дијабетесна ретинопатија']

# Rename the columns
df.rename(columns=dict(zip(df.columns, new_columns)), inplace=True)
df
```

---

### 3. Find the patient which currently has the highest glucose level. Sketch (on a plot of your choice) all of the patients glucose levels and describe the plot. Draw a conclusion on how the data behaves.

```python
# Find the patient with the highest glucose level
highest_glucose_patient = df[df['Глукоза во крв'] == df['Глукоза во крв'].max()]
print(highest_glucose_patient)

# Plot all patients' glucose levels
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Глукоза во крв'], marker='o', s=50, alpha=0.5, c='green', label='Glucose Levels')
plt.scatter(highest_glucose_patient.index, highest_glucose_patient['Глукоза во крв'], c='red', marker='o', s=100, label='Highest Glucose')
plt.xlabel('Идентификационен број на пациент')
plt.ylabel('Ниво на глукоза')
plt.title('Ниво на глукоза во крв на сите пациенти')
plt.legend()
plt.show()

# From 3 and a half to 5.7, blood glucose values are allowed, half of patients are in that range, the remaining half are 
# distributed above and below those reference values, most of them in the range of 6 to 8, a smaller proportion being above 8. 
# One patient has the highest level of glucose in the blood, and the graph is red-colored. 
# That patient has identification number 664.
```

```python
top_20_glucose_patients = df.nlargest(20, 'Глукоза во крв')

pivot_table = top_20_glucose_patients.pivot_table(values='Глукоза во крв', index=top_20_glucose_patients.index, columns='Пациент_ИБ')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlOrRd', linewidths=0.5)

plt.hlines(y=range(0, len(pivot_table)), xmin=0, xmax=len(pivot_table.columns), colors='gray', linewidths=0.5)

plt.vlines(x=range(0, len(pivot_table.columns)), ymin=0, ymax=len(pivot_table), colors='gray', linewidths=0.5)

plt.xlabel('Идентификационен број на пациент')
plt.title('Ниво на глукоза во крв на најзагрозените 20 пациенти')
plt.show()
```

### 4. Find the average number of hemoglobin A1C levels of patients per length of disease. Which duration of the disease has the highest average glucose levels? Sketch the data on a plot of your choice to best describe the data. Write a sentence or two describing how the data behaves.

```python
average_hemoglobin_per_duration = df.groupby('Број на месеци')['Хемоглобин 3 месечен'].mean()
#print(f'Average number of hemoglobin levels of patients per length of disease:\n{average_hemoglobin_per_duration} ')
highest_average_glucose_duration = average_hemoglobin_per_duration.idxmax()
print(f'The highest average glucose levels has patients with {highest_average_glucose_duration} months of duration of the disease.')

# Scatter график за односот помеѓу должината на болеста и вредноста на хемоглобинот
plt.scatter(average_hemoglobin_per_duration.index, average_hemoglobin_per_duration.values, marker='o')
plt.xlabel('Должина на болест изразена во месеци')
plt.ylabel('Просечна вредност на хемоглобинот A1C - шеќер')
plt.title('Однос помеѓу должина на болеста и вредност на хемоглобинот А1C')
plt.grid(True)
plt.show()


# The data, i.e. the graph, show that there is no relation to the length of the disease and the level of hemoglobin, 
# the image of each patient has its own representation.
```

### 5. Find the average duration of the disease for patients that are on insulin and patients that are on tablets. Compare the results. Which method that the patients use shows better results?

```python
# Calculate the average duration for patients on insulin
insulin_duration_avg = df[df['Инсулин_1/Таблети_0'] == 1]['Број на месеци'].mean()

# Calculate the average duration for patients on tablets
tablets_duration_avg = df[df['Инсулин_1/Таблети_0'] == 0]['Број на месеци'].mean()

# Round the results to two decimal places
insulin_duration_avg = round(insulin_duration_avg)
tablets_duration_avg = round(tablets_duration_avg)

print(f'Average duration for patients on insulin is {insulin_duration_avg} months.')
print(f'Average duration for patients on tablets is {tablets_duration_avg} months.')

print('  ')
if insulin_duration_avg < tablets_duration_avg:
    print("Patients using insulin show better results.")
elif insulin_duration_avg > tablets_duration_avg:
    print("Patients using tablets show better results.")
else:
    print("There is no significant difference between the two treatment methods in terms of average duration.")

```

### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python
# Define the threshold for normal A1C level
normal_A1C_threshold = 5.7

# Divide the data into normal A1C and elevated A1C groups
normal_A1C_data = df[df['Хемоглобин 3 месечен'] < normal_A1C_threshold]
elevated_A1C_data = df[df['Хемоглобин 3 месечен'] >= normal_A1C_threshold]

# Create a scatter plot to visualize the data
plt.figure(figsize=(10, 6))

# Plot the normal A1C group
plt.scatter(normal_A1C_data['Број на месеци'], normal_A1C_data['Хемоглобин 3 месечен'], color='blue', label='Normal Level')

# Plot the elevated A1C group
plt.scatter(elevated_A1C_data['Број на месеци'], elevated_A1C_data['Хемоглобин 3 месечен'], color='red', label='Elevated Level')

# Set labels and title
plt.xlabel('должина на болест изразена во месеци')
plt.ylabel('вредности на хемоглобин А1Ц')
plt.title('Ниво на хемоглобин А1Ц врз основа на должина на болест')

# Add a legend
plt.legend()

# Display the plot
plt.show()

# We see that regardless of the duration of the disease,
# we may have patients with high Levels of A1C or patients with normal A1C levels. Тhey are not in any relation.
```

```python
normal_A1C_data = df[df['Хемоглобин 3 месечен'] < normal_A1C_threshold]
elevated_A1C_data = df[df['Хемоглобин 3 месечен'] >= normal_A1C_threshold]
normal_A1C_threshold = 5.7

normal_count = len(normal_A1C_data)
elevated_count = len(elevated_A1C_data)

# Create a list of category labels
labels = ['Нормално ниво', 'Покачено ниво']

# Create a list of category counts
counts = [normal_count, elevated_count]

# Create a pie chart
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])

# Set title
plt.title('Дистрибуција на пациенти со нормално и покачено ниво на А1C')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the plot
plt.show()

# Strict division between patients, those who have normal values and those who have elevated A1C values.
```

### 7. How do factors influence the outcome whether a patient will have (or not) diabetic retinopathy? Analyse whether there is a correlation or not between diabetic retinopathy and other data in the set. Do this by separating the patients into 2 groups and analyzing the characteristics of the groups with respect to various information. Plot various data points if possible.

```python
group0 = df[df['Дијабетесна ретинопатија'] == 0]
group1 = df[df['Дијабетесна ретинопатија'] == 1]

plt.figure(figsize=(10, 5))
plt.hist(group0['Хемоглобин 3 месечен'], bins=20, alpha=0.5, label='No Retinopathy')
plt.hist(group1['Хемоглобин 3 месечен'], bins=20, alpha=0.5, label='Retinopathy')
plt.xlabel('3 месечен хемоглобин')
plt.ylabel('Времетраење на болест- изразено во месеци')
plt.legend()
plt.show()
```

```python
# On average, patients without diabetic retinopathy experience illness for a longer duration 
# of 40 months compared to those with retinopathy.
```

```python
# Pie chart for tablets and injections
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie([group0['Инсулин_1/Таблети_0'].sum(), group0['Инсулин_1/Таблети_0'].count() - group0['Инсулин_1/Таблети_0'].sum()], labels=['Tablets', 'Insulin'], autopct='%1.1f%%')
plt.title('No Retinopathy')

plt.subplot(1, 2, 2)
plt.pie([group1['Инсулин_1/Таблети_0'].sum(), group1['Инсулин_1/Таблети_0'].count() - group1['Инсулин_1/Таблети_0'].sum()], labels=['Tablets', 'Insulin'], autopct='%1.1f%%')
plt.title('Retinopathy')

plt.suptitle('Tablets vs. Insulin')

plt.show()
```

```python
# Those with diabetic retinopathy have 5% more patients on tablets and a slightly 
# smaller percentage on insulin compared to those without retinopathy. However, there is no strict correlation or rule.
# Any patient can be on tablets or insulin regardless of their retinopathy status.
```
